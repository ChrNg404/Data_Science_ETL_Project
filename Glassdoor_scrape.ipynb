```python
from splinter import Browser
from bs4 import BeautifulSoup as bs
import pandas as pd
import json
import splinter
import requests
```

```python
# Scrape Glassdoor.com
```

```python
# https://splinter.readthedocs.io/en/latest/drivers/chrome.html
!which chromedriver

executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)

# Create lists
titles = []
companies = []
locations = []
salaries = []
urls = []

for i in range(1, 31):
    # URL of 1st page of results
    if i == 1:
        url = "https://www.glassdoor.com/Job/california-data-scientist-jobs-SRCH_IL.0,10_IS2280_KO11,25.htm?minSalary=23000&includeNoSalaryJobs=false"
    # URL of 2nd page and beyond, iterate IP#
    else:
        url = f"https://www.glassdoor.com/Job/california-data-scientist-jobs-SRCH_IL.0,10_IS2280_KO11,25_IP{i}.htm?minSalary=23000&includeNoSalaryJobs=false"
    
    # request page content and create bs object
    #response = requests.get(url)
    #soup = bs(response.text,'lxml')
    browser.visit(url)
    soup = bs(browser.html, "html.parser")
    
    # On the result page
    # Scrape
    # Create list and append info
    
    try: 
        # Scrape title and url
        for x in soup.find_all('div', class_="flexbox jobTitle"):
            try:
                title = x.text
                url = "https://www.glassdoor.com" + x.a.get("href")
            # skip and fill in "na" when tag has no text or link
            except:
                title = "NaN"
            
            titles.append(title)
            urls.append(url)
            
        # Scrape company name    
        for x in soup.find_all('div', class_="flexbox empLoc"):
            try:
                company = x.text
                
            # skip and fill in "na" when tag has no text or link
            except:
                company = "NaN"
            
            companies.append(company)
            
        # Scrape location
        for x in soup.find_all('span', class_="subtle loc"):
            try:
                location = x.text
                
            # skip and fill in "na" when tag has no text or link
            except:
                location = "NaN"
            
            locations.append(location)
            
         # Scrape salary
        for x in soup.find_all('span', class_="green small"):
            try:
                salary = x.text
                
            # skip and fill in "na" when tag has no text or link
            except:
                salary = "NaN"
            
            salaries.append(salary)
            
        
    
    # skip page with no information
    except:
        print("empty page")
    
```

```python
print(len(titles))
print(len(companies))
print(len(salaries))
print(len(urls))
print(len(locations))
```

```python
glassdoor_df = pd.DataFrame({
    'Company': companies,
    'Title': titles,
    'Location': locations,
    'Salary': salaries,
    'URL': urls   
})
```

```python
# Remove duplicates from glassdoor listings
glassdoor_df.drop_duplicates(subset='URL', inplace=True)
glassdoor_df.drop_duplicates(subset=['Company','Title','Location'], inplace=True)
glassdoor_df = glassdoor_df.reset_index(drop=True)
```

```python
# https://splinter.readthedocs.io/en/latest/drivers/chrome.html
# open ChromeDriver again in case it was closed
!which chromedriver

executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)

# Define a list to hold descriptions
descs = []
# Counter
i = 0

for url in glassdoor_df['URL'].tolist():
    # Open this URL
    browser.visit(url)
    # Create soup object for this URL
    
    soup=bs(browser.html, "html5lib") 
    # Scrap for description if exists 
    try:
        soup.find('div', class_="jobDescriptionContent desc")
        
        try:
            desc = soup.find('div', class_="jobDescriptionContent desc").text
            print(i)
        # leave blank in no text in section
        except:
            desc = " "          
    
    # Otherwise leave blank
    except AttributeError as e:
        print("error")
        desc = " "
    
    descs.append(desc)
    i += 1
```

```python
# Add Description to glassdoor_df and clean up
glassdoor_df["Description"] = descs
glassdoor_df["Description"] = glassdoor_df["Description"].str.strip('\n')

glassdoor_df["Company"] = glassdoor_df["Company"].str.split('–', expand=True)[0]

# Add column "Site"
glassdoor_df["Site"] = "Glassdoor"
# Add column "Keyword"
glassdoor_df["Keyword"] = "Data Scientist"
# Split state and city
glassdoor_df["City"] =  glassdoor_df["Location"].str.split(',', expand=True)[0]
glassdoor_df["State"] = glassdoor_df["Location"].str.split(',', expand=True)[1]
```
