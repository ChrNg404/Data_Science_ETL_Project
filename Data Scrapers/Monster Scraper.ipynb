## Scraping Monster

```python
# URL of the pages to be scraped
url = 'https://www.monster.com/jobs/search/?q=Data-analyst&where=California&intcid=skr_navigation_nhpso_searchMain&jobid=203706284'
#browser
browser.visit(url)
#define the html
time.sleep(1)
#define the soup
html=browser.html
soup=bs(html,'html.parser')
# Retrieve page with requests module
response = requests.get(url)

#Create BeautifulSOup object; parse with 'lxml'

soup = bs(response.text,'lxml')
```

```python
# Retrieve data from Nasa website
titles=soup.find_all('h2',class_='title')
title_list = []
companies=soup.find_all('div',class_='company')
company_list = []
locations= soup.find_all('a', class_='name')
location_list = []
joburls = soup.find_all('h2', class_='title')
url_list = []
#print(title)
#print(company)
print(locations)
```

```python
for title in titles:
    title_list.append(title.text)
    url_list.append(title.a.get('href'))
    print(title.text)
    print(title.a.get('href'))
```

```python
for location in locations:
    location_list.append(location.text)
    print(location.text)  
```

```python
for company in companies:
    company_list.append(company.text)
    print(company.text)
```

```python
post = {
    'title': title_list,
    'location':location_list,
    'company': company_list,
    'url':url_list
}
print(post)
```

```python
results = soup.find_all('div', class_="summary")
#Loop through reults to retrieve each piece I need

for result in results:
    title = result.find('h2',class_='title').text
    company = result.find('div',class_='company').text
    location = result.find('a', class_='name').text
    joburl=result.find('h2', class_='title').a.get('href')
    print(title)
    print(company)
    print(location)
    print(joburl)
    
    post = {
        'title':title,
        'company':company,
        'location':location,
        'url':joburl
    }
    
    collection.insert_one(post)
```
